### 2.使用ollama的python接口

```python
# !pip install ollama
import ollama
from IPython.display import Markdown
```

```python


msg = [
  {
    'role': 'user',
    'content': 'Pytorch是什么，它主要用于什么方面？',
  },
]

%time response = ollama.chat(model='qwen2:7b-instruct', messages=msg)
```

```python
response
```

```python
# 显示markdown内容
Markdown(response["message"]["content"])
```
